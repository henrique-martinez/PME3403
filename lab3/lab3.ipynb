## f) Fazer um modelo simples da suspensão, que explique matematicamente a eficácia no isolamento da vibração do gabinete.

$$(M-m) \ddot x  + m \ddot{ ( \overline{x + e \Omega^2 sin(\Omega t)})} + c \dot x = -\frac{Mg}{L} x $$
$$M \ddot x + c \dot x + \frac{Mg}{L} x = me \Omega^2 sin(\Omega t)$$

Assim temos os seguintes termos:

$$
\begin{cases}
    M_{eq} = M \\
    K_{eq} = \frac{Mg}{L} \\
    C_{eq} = c \; (na \;\; hipótese \;\; inicial \;= \; 0) \\
    F_C = me \Omega^2
\end{cases}
$$

Assim podemos calcular os seguintes parâmetros:

$$
\begin{cases}
    \omega= \sqrt{g/L_{eq}} \\
    r= \Omega / \omega \\
    X_p = \frac{F_C/k}{\sqrt{(1+r^2)^2 + (2 \xi r )^2}} = \frac{meL \Omega^2/Mg}{\sqrt{(1+r^2)^2 + (2 \xi r )^2}} \\
    \frac{F_r}{F_C} = \frac{Mg \dot X_p / L}{me \Omega^2}
\end{cases}
$$

Agora, substituindo pelos valores estimados. Incicialmente, para estimar a frequência natural do tambor suspenso, um registro foi feito com o celular aderido no tambor. O espectro de frequência abaixo mostra que:
$$ f_n = 1,2Hz \Rightarrow \omega_n = 2 \pi f_n = 7,54 \; rad/s $$

Assim podemos calcular o $L_{eq}$, usando $g=9,79 \; m/s^2$:

$$\omega = \sqrt{\frac{g}{L_{eq}}} \Rightarrow L_{eq} = \frac{g}{\omega^2} =\frac{9,79}{7,54^2} = 0,172 m$$

A veocidade de operação da máquina é $\Omega = 1000rpm=954,93 rad/s$. A massa estimada do sistema é $M=30kg$. A massa desbalanceada é $m=500g$, com raio de desbalanceamento $e=18cm$. Assim podemos calcular a transmissibilidade de força:

$$r = \frac{\Omega}{omega}= 954,93/7,54 = 126,65 $$

$$X_p = \frac{meL \Omega^2/Mg}{\sqrt{(1+r^2)^2 + (2 \xi r )^2}} = \frac{0,5 \cdot 0,18 \cdot 0,172 \cdot 954,93^2/(30 \cdot 9,79)}{\sqrt{(1+126,65^2)^2}} = \frac{0,5 \cdot 0,18 \cdot 0,172 \cdot 954,93^2/(30 \cdot 9,79)}{1+126,65^2} = 0,029m $$

$$\frac{F_r}{F_C} = \frac{Mg \cdot X_p / L}{me \Omega^2} = \frac{30 \cdot 9,79 \cdot 0,029 / 0,172}{0,5 \cdot 0,18 \cdot 954,93^2} = 6,03 \cdot 10^{-4}$$
